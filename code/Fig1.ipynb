# Plot EasyOcean sections on top of topography, color representing numbers of occupations

1) only plot EasyOcean sections that were occupied more than once
2) use approximate section locations from median

```python
%matplotlib inline
import numpy as np
import xarray as xr
import scipy.io as sio
import os
import matplotlib.pyplot as plt

import topo as topo
```

```python
def plot_sec(sec, method, datafilepath, ax, vmin, vmax, cmap, fontsize=15):
    """method = 1: lon section
       method = 0, lat section """
    # station_loc[occupation, lon/lat, station]

    loc = np.zeros(len(sec),) + np.nan
    count = 0
    for acquisition in sec: #os.listdir(datafilepath):#
        if not acquisition.startswith('.'):
            acqui_path = datafilepath + acquisition + '/'
            print(10*"-", acquisition, 10*"-")
            for filename_ in os.listdir(acqui_path):
                if (filename_.endswith(".mat")):  # mat data
                    data0 = sio.loadmat(acqui_path + filename_,
                                        squeeze_me=True, struct_as_record=False)
                    occup_count = np.zeros(len(data0['ll_grid']),)
                    station_loc = np.zeros((len(data0['D_pr']), 2, 1000)) + np.nan
                    for i in range(len(data0['D_pr'])):
                        for j in range(len(data0['D_pr'][i].Station)):
                            station_loc[i,0,j] = data0['D_pr'][i].Station[j].Lon
                            station_loc[i,1,j] = data0['D_pr'][i].Station[j].Lat
                        if method:
                            occup_count[np.where((data0['ll_grid']>=station_loc[i,1,:][~np.isnan(station_loc[i,1,:])][0]) & (data0['ll_grid']<=station_loc[i,1,:][~np.isnan(station_loc[i,1,:])][-1]))]+=1
                        else:
                            if (data0['ll_grid'][0]<0) & (data0['ll_grid'][-1]>0): # cross 0 degree
                                lon_ = station_loc[i,0,:][~np.isnan(station_loc[i,0,:])]
                                lon_[lon_>180]-=360
                                occup_count[np.where((data0['ll_grid']>=lon_[0]) & (data0['ll_grid']<=lon_[-1]))]+=1
                            else:
                                occup_count[np.where((data0['ll_grid']>=station_loc[i,0,:][~np.isnan(station_loc[i,0,:])][0]) & (data0['ll_grid']<=station_loc[i,0,:][~np.isnan(station_loc[i,0,:])][-1]))]+=1
                    if method:
                        loc[count] = np.nanmedian(station_loc[:,0,:])
                        sc = ax.scatter(np.ones(data0['ll_grid'].shape)*np.nanmedian(station_loc[:,0,:]), data0['ll_grid'], c=occup_count, vmin=vmin, vmax=vmax, cmap=cmap)
                        if np.nanmin(data0['ll_grid'])<30:
                            ax.text(np.nanmedian(station_loc[:,0,:]), data0['ll_grid'][0], acquisition, 
                                    fontsize=fontsize, horizontalalignment='center', verticalalignment='top', rotation='vertical')
                        else:
                            ax.text(np.nanmedian(station_loc[:,0,:]), data0['ll_grid'][-1], acquisition, 
                                    fontsize=fontsize, horizontalalignment='center', verticalalignment='bottom', rotation='vertical')                            
                    else:
                        loc[count] = np.nanmedian(station_loc[:,1,:])
                        lon = data0['ll_grid'].copy()
                        lon[lon<0]+=360
                        sc = ax.scatter(lon, np.ones(data0['ll_grid'].shape)*np.nanmedian(station_loc[:,1,:]), c=occup_count, vmin=vmin, vmax=vmax, cmap=cmap)
                        ax.text(lon[0], np.nanmedian(station_loc[:,1,:]), acquisition, 
                                fontsize=fontsize, horizontalalignment='center', verticalalignment='top')
        count += 1
    return sc, loc
```
